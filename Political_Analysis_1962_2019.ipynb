# HANDLE MISSING VALUES AND TYPE CONVERSIONS





```python
# Fill missing Party_Type with "Unknown"

df['Party_Type'].fillna('Unknown' ,inplace=True)

# Fill missing categorical columns with "Unknown"
df['Party_Name'].fillna('Unknown', inplace=True)
df['Frequent_Abbreviation'].fillna('Unknown', inplace=True)
df['Last_Abbreviation'].fillna('Unknown', inplace=True)
df['Abbreviations'].fillna('Unknown', inplace=True)

# Fill missing numeric columns with the median value
df['Party_ID'].fillna(df['Party_ID'].median(), inplace=True)
df['Start_Year'].fillna(df['Start_Year'].median(), inplace=True)
df['Last_Year'].fillna(df['Last_Year'].median(), inplace=True)

 # Data Type Conversion
df['Start_Year'] = df['Start_Year'].astype(int)
df['Last_Year'] = df['Last_Year'].astype(int)

#  Consistency Checks
# Ensure year consistency
df = df[df['Start_Year']<= df['Last_Year']]

# Save the cleaned file in Colab environment
cleaned_file_path = '/content/cleaned_1962_2021.csv'
df.to_csv(cleaned_file_path, index=False)

# Verification of the cleaned data
print(f"Cleaned data saved to: {cleaned_file_path}")

#Verification of the cleaned data
print(df.info())
print(df.head())

```

# **Explorative Data Analysis**

**Top 5 Party Participation Trend**

```python
def party_participation_trends(df, top_n=5, exclude_parties=["NA's", 'None of the Above', 'Independent', 'Expanded Party name not released by the ECI']):
    # Calculate total participation from each party excluding specified parties
    party_total = df.loc[~df['Party_Name'].isin(exclude_parties), 'Party_Name'].value_counts().head(top_n)
    top_parties = party_total.index

    # Filter dataframe to include the top parties and exclude specified parties
    filtered_df = df[df['Party_Name'].isin(top_parties)]

    # Group by Start_Year and Party_Name, calculate size, and unstack to pivot table
    yearly_counts = filtered_df.groupby(['Start_Year', 'Party_Name']).size().unstack(fill_value=0)

    # Plotting the bar plot
    fig, ax = plt.subplots(figsize=(15, 7))
    yearly_counts.plot(kind='bar', stacked=True, ax=ax)

    # Adding exact numbers on the bars
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f', label_type='center')

    plt.title('Top {} Political Party Participation Trends (1962-2019)'.format(top_n))
    plt.xlabel('Year')
    plt.ylabel('Number of Events')
    plt.legend(title='Party', loc='upper left')
    plt.tight_layout()
    plt.show()

# Example usage: Display top 5 political party participation trends, excluding 'NA' and 'None of the above'
party_participation_trends(df, top_n=5, exclude_parties=["NA's", 'None of the Above', 'Independent', 'Expanded Party name not released by the ECI'])
```

Inference : The particpation of Communist PArty of India has reduced from 1962 when it was the highest, 28  has now reduced to 4 by 2014.

**Female Representation**

```python
def female_representation(df):
    # Define party names to exclude
    exclude_parties = ["NA's", 'None of the Above', 'Independent', 'Expanded Party name not released by the ECI']

    # Filter out excluded parties
    filtered_df = df[~df['Party_Name'].isin(exclude_parties)]

    # Calculate yearly counts for female candidates contested and represented
    yearly_female_counts = filtered_df.groupby(['Start_Year', 'Party_Name'])[['Females_Contested', 'Females_Represented']].sum()

    # Aggregate by Start_Year and sum Females_Contested and Females_Represented
    yearly_female_counts = yearly_female_counts.reset_index()
    party_totals = yearly_female_counts.groupby('Party_Name')[['Females_Contested', 'Females_Represented']].sum()

    # Sort by total Females_Contested descending and take top 5 parties
    top_parties = party_totals.sort_values(by='Females_Contested', ascending=False).head(5).index

    # Filter DataFrame to include only top parties
    yearly_female_counts_top = yearly_female_counts[yearly_female_counts['Party_Name'].isin(top_parties)]

    # Sum the counts for each party across years
    party_female_counts = yearly_female_counts_top.groupby('Party_Name')[['Females_Contested', 'Females_Represented']].sum().loc[top_parties]

    # Plotting
    fig, ax = plt.subplots(figsize=(12, 7))
    bar_width = 0.4
    positions = np.arange(len(top_parties))

    bars_contested = ax.bar(positions - bar_width/2, party_female_counts['Females_Contested'], width=bar_width, label='Females Contested')
    bars_represented = ax.bar(positions + bar_width/2, party_female_counts['Females_Represented'], width=bar_width, label='Females Represented')

    # Adding exact numbers on the bars
    ax.bar_label(bars_contested, fmt='%.0f')
    ax.bar_label(bars_represented, fmt='%.0f')

    ax.set_title('Top 5 Parties: Female Candidates Contested and Represented (1962-2019)')
    ax.set_xlabel('Party Name')
    ax.set_ylabel('Number of Female Candidates')
    ax.set_xticks(positions)
    ax.set_xticklabels(top_parties, rotation=45, ha='right')
    ax.legend(title='Representation Type', bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()

# Call the function with your DataFrame `df`
female_representation(df)
```

Inference : From 1962 -2019 The Congress Party has been the highest in terms of Female Candidate Representation which is 4297 and the highest in terms of Candidate Contested which is 1642 . On the other hand Communist Marxist PArty of India has been the lowest in tersm of contested and BSP has been the lowest in terms of represented






**Success Rate in Reserved Seats**

```python
# Define party names to exclude
exclude_parties = ["NA's", 'None of the Above', 'Independent', 'Expanded Party name not released by the ECI']

# Filter out excluded parties
filtered_df = df[~df['Party_Name'].isin(exclude_parties)]

# Success rate in SC/ST seats by party
reserved_seats = filtered_df.groupby('Party_Name')[['SC_Seats_Contested', 'SC_Seats_Represented', 'ST_Seats_Contested', 'ST_Seats_Represented']].sum()

# Calculate success rates
reserved_seats['SC_Success_Rate'] = reserved_seats['SC_Seats_Represented'] / reserved_seats['SC_Seats_Contested']
reserved_seats['ST_Success_Rate'] = reserved_seats['ST_Seats_Represented'] / reserved_seats['ST_Seats_Contested']

# Sort by SC and ST seats contested and take the top 5 parties
top_parties = reserved_seats.sort_values(by=['SC_Seats_Contested', 'ST_Seats_Contested'], ascending=False).head(5)

# Plotting
ax = top_parties[['SC_Success_Rate', 'ST_Success_Rate']].plot(kind='bar', figsize=(14, 7), colormap='tab20')

# Adding exact numbers on the bars as percentages
for container in ax.containers:
    labels = [f'{v * 100:.1f}%' for v in container.datavalues]
    ax.bar_label(container, labels=labels)

# Customizing the plot for better visibility
plt.title('Success Rates in Reserved Seats by Party')
plt.xlabel('Party')
plt.ylabel('Success Rate')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Success Rate', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Display the plot
plt.show()
```

Inference : Communist Marxist Party of India has had the highest SC success rate whereas on the other hand, Congess has had the highest ST success rate

**Performance Analysis**

```python


# Calculate success rates
party_success = df.groupby('Party_Name')[['Candidates_Contested', 'Candidates_Represented']].sum()
party_success['Success_Rate'] = (party_success['Candidates_Represented'] / party_success['Candidates_Contested']) * 100  # Convert to percentage

# Get top 5 parties with highest success rates
top_success_parties = party_success.sort_values('Success_Rate', ascending=False).head(5)

# Define a custom color palette (adjust colors as needed)
custom_palette = sns.color_palette('husl', n_colors=10)

# Plotting
plt.figure(figsize=(14, 7))
bars = plt.bar(top_success_parties.index, top_success_parties['Success_Rate'], color=custom_palette)

# Adding percentages on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom', fontsize=8)

plt.title('Top 5 Parties with Highest Success Rates')
plt.xlabel('Party')
plt.ylabel('Success Rate (%)')
plt.xticks(rotation=45)
plt.ylim(0, 100)  # Ensure y-axis starts from 0 and goes up to 100
plt.tight_layout()
plt.show()
```

**BiPoll Contest**

```python


# Calculate BiPoll participation
bipoll_analysis = df.groupby('Party_Name')['BiPoll_Contested'].sum().sort_values(ascending=False).head(5)

# Define a custom color palette (adjust colors as needed)
custom_palette = sns.color_palette('husl', n_colors=5)

# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
bars = plt.bar(bipoll_analysis.index, bipoll_analysis.values, color=custom_palette)

# Adding number of BiPoll contests on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Parties in BiPoll Contests')
plt.xlabel('Party')
plt.ylabel('Number of BiPoll Contests')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
